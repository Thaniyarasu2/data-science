```python
d.groupby("gender")[["math score", "reading score", "writing score"]].mean().plot(kind="bar")
plt.title("gender vs score")
plt.xlabel("gender")
plt.ylabel("score")
plt.legend()
plt.show()
```

```python
plt.hist(d["math score"],bins=20)
plt.title("distribution of math score")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show()
```

```python
#scatter chat
plt.scatter(d["math score"],d["reading score"])
plt.title("distribution of math score")
plt.xlabel("score")
plt.ylabel("frequency")
plt.show()
```

```python
#line plot
avg_s=d[["math score", "reading score","writing score"]].mean()
avg_s
```

```python
x=avg_s.index
y=avg_s.values
```

```python
plt.plot(x,y)
plt.title("line")
plt.xlabel("cato")
plt.ylabel("scoring")
plt.show()
```
