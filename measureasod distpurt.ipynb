```python
import numpy as np
```

```python
normal_data=np.random.normal(loc=50,scale=3,size=1000)
normal_data
```

```python
o=np.array([100,150,200,300,400])
o_data=np.concatenate([normal_data,o])
o_data
```

```python
nm=np.mean(normal_data)
om=np.mean(o_data)
print(nm)
print(om)
```
