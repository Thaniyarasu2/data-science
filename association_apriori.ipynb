```python
import pandas as pd
```

```python
d=pd.read_csv("market.csv")
d
```

```python
head=d.columns[0].split(';')
head
```

```python
s1=d.iloc[:,0].str.split(';',expand=True)
s1
```

```python
s1.columns=head
s1
```

```python
s1.isnull().sum()
```

```python
s1=s1.astype(int)
s1
```

```python
s1.shape
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
fi=apriori(s1,min_support=0.1,use_colnames=True)
print(fi)
```

```python
n=association_rules(fi,metric="confidence",min_threshold=0.3)
print(n)
```

```python
print(n[['antecedents','consequents','support','confidence','lift']].sort_values(by="lift",ascending=False))



```

```python
item_frequencies=s1.sum().sort_values(ascending=False)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.barplot(x=item_frequencies.head(10).values,y=item_frequencies.head(10).index)
plt.title("Top 10 frequently items")
plt.xlabel("Frequently")
plt.ylabel("items")
plt.show()
```
