```python
x=s[["danceability","tempo","loudness","Valence","energy"]]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x=ss.fit_transform(x)
x
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(n_neighbors=6,metric="euclidean")
model.fit(x)
```

```python
#s["song_name"].unique()
```

```python
def rec(song):
    s_index = s[s["song_name"] == song].index[0]
    
    query_point = s.loc[[s_index], x].values
   
    distances, indices = model.kneighbors(query_point)
    
    return s.iloc[indices[0]]["song_name"]

print(rec("Andhi Mazhai Pozhigirathu"))
```
